```python
import numpy as np
import plotly.subplots as sp
import plotly.graph_objs as go

# Gere dados aleat√≥rios
num_samples = 50
x = np.arange(num_samples)
y_bar = np.random.rand(num_samples)
y_cascade = np.random.randn(num_samples)
y_line = np.cumsum(np.random.randn(num_samples))
y_scatter = np.random.rand(num_samples)

# Crie o layout do dashboard
fig = sp.make_subplots(rows=2, cols=2, subplot_titles=("Gr√°fico de Barras", "Gr√°fico de Cascata", "Gr√°fico de Linhas", "Gr√°fico de Dispers√£o"))

# Adicione os gr√°ficos ao layout
fig.add_trace(go.Bar(x=x, y=y_bar), row=1, col=1)
fig.add_trace(go.Waterfall(x=x, y=y_cascade), row=1, col=2)
fig.add_trace(go.Scatter(x=x, y=y_line, mode='lines'), row=2, col=1)
fig.add_trace(go.Scatter(x=x, y=y_scatter, mode='markers'), row=2, col=2)

# Atualize o t√≠tulo e o layout
fig.update_layout(
    title="<b>Dashboard com N√∫meros Aleat√≥rios com Python - Cleison Lima</b><br>Vers√£o 1.0",
    title_x=0.5,  # Centralize o t√≠tulo horizontalmente
    paper_bgcolor='black',  # Defina a cor de fundo para preto
    plot_bgcolor='black',  # Defina a cor de fundo dos gr√°ficos para preto
    font=dict(color='white')  # Defina a cor da fonte para branco
)

# Exiba o dashboard
fig.show()

```

```python
import numpy as np
import plotly.subplots as sp
import plotly.graph_objs as go
import plotly.io as pio

# Gere dados aleat√≥rios
num_samples = 50
x = np.arange(num_samples)
y_bar = np.random.rand(num_samples)
y_cascade = np.random.randn(num_samples)
y_line = np.cumsum(np.random.randn(num_samples))
y_scatter = np.random.rand(num_samples)

# Crie o layout do dashboard
fig = sp.make_subplots(rows=2, cols=2, subplot_titles=("Gr√°fico de Barras", "Gr√°fico de Cascata", "Gr√°fico de Linhas", "Gr√°fico de Dispers√£o"))

# Adicione os gr√°ficos ao layout
fig.add_trace(go.Bar(x=x, y=y_bar), row=1, col=1)
fig.add_trace(go.Waterfall(x=x, y=y_cascade), row=1, col=2)
fig.add_trace(go.Scatter(x=x, y=y_line, mode='lines'), row=2, col=1)
fig.add_trace(go.Scatter(x=x, y=y_scatter, mode='markers'), row=2, col=2)

# Atualize o t√≠tulo e o layout
fig.update_layout(
    title="<b>Dashboard com N√∫meros Aleat√≥rios com Python - Cleison Lima</b><br>Vers√£o 1.0",
    title_x=0.5,  # Centralize o t√≠tulo horizontalmente
    paper_bgcolor='black',  # Defina a cor de fundo para preto
    plot_bgcolor='black',  # Defina a cor de fundo dos gr√°ficos para preto
    font=dict(color='white')  # Defina a cor da fonte para branco
)

# Salve o dashboard como arquivo HTML
html_file = "dashboard.html"
pio.write_html(fig, file=html_file, auto_open=True)
```
