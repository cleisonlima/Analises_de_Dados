```python
import pandas as pd
import pandera as pa
```

pip install pandera # Instalação de pacotes pandera

```python
df = pd.read_csv("/content/drive/MyDrive/ocorrencias.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
```

```python
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int), # Verfica se os dados são do tipo initeiros
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2)), # Informamos o mínimo ou o máximo de caracteres
        "ocorrencia_aerodromo": pa.Column(pa.String),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'),nullable=True), # Encontrado valores nulos na coluna
        "total_recomendacoes": pa.Column(pa.Int)
    
    }
)
```

```python
schema.validate(df)
```
