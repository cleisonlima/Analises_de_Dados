# Rank dos melhores jogos para Android
 Os dados foram importados a partir do link: https://www.kaggle.com/dhruvildave/top-play-store-games.


```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("android-games.csv")

#Tratamento e limpeza dos dados
data = data.dropna(how="all", axis=1) #Elimina todoas as colunas vazias
data = data.dropna(how="any", axis=0) #Eliminina todas as linhas vazias, podemos deixar ()vazio também
#print(data.info()) #Visualizar exclusão das linhas e colunas vazias
```

```python
data.head()
```

```python
plt.figure(figsize=(15,7))

plt.title("RANK DOS JOGOS MAIS BEM AVALIADOS NO ANDROID - TOP 5")

jogos = ('Garena Free Fire - The Cobra', 
         'PUBG MOBILE: Graffiti Prank', 
         'Mobile Legends: Bang Bang', 
         'Brawl Stars', 
         'Sniper 3D: Fun Free Online FPS Shooting Game')
top = [80678661,35971961,25836869,17181659,14237554]

plt.bar(jogos, top)
plt.xlabel('JOGOS')
plt.ylabel('TOPO')
plt.show()
```

```python
plt.figure(figsize=(15,7))

plt.title('JOGOS QUE RECEBERAM MAIS QUANTIDADES DE 5 ESTRELAS')

jogos = ('Garena Free Fire - The Cobra', 
         'PUBG MOBILE: Graffiti Prank', 
         'Mobile Legends: Bang Bang', 
         'Brawl Stars', 
         'Sniper 3D: Fun Free Online FPS Shooting Game')

top = [61935712, 26670566, 17850942, 12493668, 9657878]
colors = ['CornflowerBlue', 'green', 'orange', 'coral', 'red']
patches, texts, autotexts = plt.pie(top, colors=colors, autopct='%1.1f%%',
startangle=90)
plt.legend(patches, jogos, loc="lower right")
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(15,7))

plt.title('GRÁFICO DE LINHAS PARA ANÁLISE DE CRESCIMENTO')

jogos = ('Garena Free Fire - The Cobra', 
         'PUBG MOBILE: Graffiti Prank', 
         'Mobile Legends: Bang Bang', 
         'Brawl Stars', 
         'Sniper 3D: Fun Free Online FPS Shooting Game')

trintadias = [2.9, 2.0, 1.6, 4.1, 0.8]
sessentadias = [7.9, 3.1, 3.3, 6.6, 1.8 ]
plt.plot(jogos, trintadias, label='Crescimento em 30 dias')
plt.plot(jogos, sessentadias, label='Crescimento em 60 dias')
plt.legend()
```
